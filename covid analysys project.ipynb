```python
df=df.filter(items=['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
```

```python
df
```

```python
df.duplicated().sum()
```

```python
new_df=df.dropna(subset=['continent'])
```

```python
new_df.isnull().sum()
```

```python
new_df['total_deaths'] = new_df['total_deaths'].fillna(0)
new_df['gdp_per_capita'] = new_df['gdp_per_capita'].fillna(0)
new_df['human_development_index'] = new_df['human_development_index'].fillna(0)
```

```python
new_df.isnull().sum()
```

```python
new_df['date'] = pd.to_datetime(new_df['date'])
```

```python
new_df
```

```python
new_df['month'] = pd.DatetimeIndex(new_df['date']).month
```

```python
new_df
```

```python
a=new_df.groupby('continent').max()
df_groupby=a.reset_index()
```

```python
df_groupby
```

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
```

```python
df_groupby
```

```python
sns.distplot(df_groupby.gdp_per_capita)
```

```python
plt.scatter(df_groupby['total_cases'],df_groupby['gdp_per_capita'])
```

```python
sns.pairplot(df_groupby)
```

```python
sns.catplot(x="continent", y="total_cases", data=df_groupby ,kind='bar')
```
